```python
import osmnx as ox
import matplotlib.pyplot as plt
```

# **SÃ£o Paulo**

```python
cidade_nome_sp = "sp"
cidade_grafo_sp = "Sao Paulo, SP, Brazil"

data_sp = "27_06_2022"
```

```python
G_sp = ox.io.load_graphml(filepath=f"../data/{cidade_nome_sp}_{data_sp}.graphml")
```

### Plot normal

```python
fig, ax = ox.plot_graph(G_sp)
```

### Plot interativo

```python
%matplotlib tk

fig, ax = ox.plot_graph(G_sp)
plt.show()
```

```python
for node in G_sp.nodes(data=True):
    print(node)
    break
```
